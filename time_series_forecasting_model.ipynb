```python
# Load data simulasi
df = pd.read_csv('/content/pt_nuansa_mart_sales_simulation.csv')
df['tanggal'] = pd.to_datetime(df['tanggal'])

# Cek data unik
print("Kategori produk:", df['kategori'].unique())
print("Kampanye:", df['kampanye'].unique())
print("Cuaca:", df['cuaca'].unique())
```

```python
def forecast_kategori(df, kategori, prediksi_hari=90):
    df_kat = df[df['kategori'] == kategori][['tanggal', 'penjualan']]
    df_kat = df_kat.rename(columns={'tanggal': 'ds', 'penjualan': 'y'})

    model = Prophet()
    model.fit(df_kat)

    future = model.make_future_dataframe(periods=prediksi_hari)
    forecast = model.predict(future)

    # Plot hasil forecast
    fig1 = model.plot(forecast)
    plt.title(f'ðŸ“ˆ Prediksi Penjualan - {kategori}')
    plt.xlabel("Tanggal")
    plt.ylabel("Penjualan")
    plt.show()

    # Plot komponen musiman
    fig2 = model.plot_components(forecast)
    plt.show()

    return forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
from IPython.display import display, Markdown

for kategori in df['kategori'].unique():
    display(Markdown(f"### ðŸ“Š Kategori: **{kategori.capitalize()}**"))
    forecast_df = forecast_kategori(df, kategori)

    # Tampilkan hanya 5 prediksi terakhir
    display(Markdown("**ðŸ“… 5 Prediksi Terakhir:**"))
    display(forecast_df.tail(5))

    print("\n" + "-"*80 + "\n")
```

```python
df['bulan'] = df['tanggal'].dt.month
df['tahun'] = df['tanggal'].dt.year

plt.figure(figsize=(10, 6))
pivot = df.pivot_table(index='bulan', columns='kategori', values='penjualan', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("ðŸ”† Rata-rata Penjualan Bulanan per Kategori")
plt.show()
```
